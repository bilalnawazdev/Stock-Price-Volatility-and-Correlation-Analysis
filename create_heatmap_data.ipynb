```python
import pandas as pd
import yfinance as yf
from datetime import datetime
```

```python
tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']
start_date = '2018-01-01'
end_date = datetime.now().strftime('%Y-%m-%d')
raw_data = yf.download(tickers, start=start_date, end=end_date)


try:
    adj_close_data = raw_data['Adj Close']
    print("Successfully found 'Adj Close' column.")
except KeyError:
    adj_close_data = raw_data['Close']
    print("'Adj Close' not found. Using 'Close' column instead.")


daily_returns = adj_close_data.pct_change().dropna()


correlation_matrix = daily_returns.corr()

correlation_matrix.index.name = None
correlation_matrix.columns.name = None


heatmap_data = correlation_matrix.stack().reset_index()
heatmap_data.columns = ['Ticker_1', 'Ticker_2', 'Correlation']

heatmap_data.to_csv('correlation_heatmap_data.csv', index=False)

print(heatmap_data.head())
```
